```python
df = pd.read_csv('/home/tobamo/analize/project-tobamo/results/megan6_results_combined_add_nr_taxa.csv')
```

```python
megan_non_virus = df[~df.megan_tax.str.contains('Virus')]
```

```python
megan_non_virus_pident = megan_non_virus[megan_non_virus.nr_pident > 50]
```

```python
non_virus = megan_non_virus_pident.groupby('qseqid').filter(lambda x: ~(x['nr_tax'].str.contains('Viruses').all()))
```
